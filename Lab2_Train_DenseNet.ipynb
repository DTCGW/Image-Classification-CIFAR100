```python
import re
import matplotlib.pyplot as plt

def show_dense_accuracy_plot_from_file(filepath):
    with open(filepath, 'r') as f:
        log_text = f.read()

    # Trích xuất Train Accuracy
    train_acc = [float(x) for x in re.findall(r"\[DenseNet121\] Epoch \d+: Train Loss=[\d\.]+, Train Acc=(\d+\.\d+)%", log_text)]

    # Trích xuất Test Accuracy
    val_acc = [float(x) for x in re.findall(r"\[DenseNet121\] Test Accuracy: (\d+\.\d+)%", log_text)]

    # Lấy epoch bắt đầu từ dòng đầu tiên
    match = re.search(r"\[DenseNet121\] Epoch (\d+):", log_text)
    start_epoch = int(match.group(1)) if match else 1
    epochs = list(range(start_epoch, start_epoch + len(train_acc)))

    # Vẽ biểu đồ
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_acc, marker='o', label='Train Accuracy')
    plt.plot(epochs, val_acc, marker='s', label='Validation Accuracy')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.title("DenseNet121: Train and Validation Accuracy per Epoch")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
show_dense_accuracy_plot_from_file("/content/drive/MyDrive/Master/CS2310.CH190/Lab2/Image-Classification-CIFAR100/training_logs/Lab2_Train_DenseNet.txt")
```
