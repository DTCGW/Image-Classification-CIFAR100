```python
import re
import matplotlib.pyplot as plt

def show_train_val_accuracy_from_log(filepath):
    with open(filepath, 'r') as f:
        log_text = f.read()

    # Trích xuất train accuracy từ dòng "Epoch [xx/xx], Loss: ..., Train Acc: xx.xx%"
    train_acc = [float(x) for x in re.findall(r"Epoch \[\d+/\d+\], Loss: [\d\.]+, Train Acc: (\d+\.\d+)%", log_text)]

    # Trích xuất test (validation) accuracy từ dòng "Test Accuracy: xx.xx%"
    val_acc = [float(x) for x in re.findall(r"Test Accuracy: (\d+\.\d+)%", log_text)]

    # Epoch bắt đầu (ví dụ: từ Epoch 86)
    match = re.search(r"Epoch \[(\d+)/\d+\]", log_text)
    start_epoch = int(match.group(1)) if match else 1
    epochs = list(range(start_epoch, start_epoch + len(train_acc)))

    # Vẽ biểu đồ
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_acc, marker='o', label='Train Accuracy')
    plt.plot(epochs, val_acc, marker='s', label='Validation Accuracy')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.title("Train and Validation Accuracy per Epoch")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
show_train_val_accuracy_from_log("/content/drive/MyDrive/Master/CS2310.CH190/Lab2/Image-Classification-CIFAR100/training_logs/Lab2_Train_ResNet.txt")
```
